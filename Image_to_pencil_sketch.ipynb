**Generate Pencil Sketch from image using OpenCV by doing image manipulation**

Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

Reading and printing images using opencv

```python
img = cv2.imread('beyblade.jpg', cv2.IMREAD_UNCHANGED)
cv2_imshow(img)
```

printing the images using matplotlib to increase efficiency and reduce hassles

```python
plt.imshow(img)
plt.axis(False)
plt.show()
```

Converting BGR to RGB

```python
#1 Using cvtColor method of OpenCV.
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.show()
```

Convert Image to a Pencil Sketch

Step 1: Convert to Grey Image

```python
#Using cvtColor function of OpenCV.
grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_img)
```

Step 2: Invert Image

```python
#Inverting the image can be done by either of the following methods
# invert_img=cv2.bitwise_not(grey_img) #1 method - cv2.bitwise_not()
invert_img=255-grey_img #2 method 2 by substracting from 255
```

```python
plt.imshow(invert_img)
```

Step 3: Blur image

```python
#Apply Gaussian blur to the image. The second argument to the function is the kernel size, if should be a pair of odd numbers.Larger the kernel size, more blurred the image will be and it will lose its subtle features.
blur_img=cv2.GaussianBlur(invert_img, (111,111),0)
plt.imshow(blur_img)
```

Step 4: Invert Blurred Image

```python
# Repeat step 2 to invert images using either of two below methods
invblur_img=cv2.bitwise_not(blur_img)
#invblur_img=255-blur_img
plt.imshow(invblur_img)
```

Step 5: Sketch

```python
# The sketch can be obtained by performing bit-wise division between the grayscale image and the inverted-blurred image.
sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)
plt.imshow(sketch_img)
```

Original Image vs Sketch

```python
plt.figure(figsize=(14,8))

#original image
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(RGB_img)
plt.axis('off')

#sketch 
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
